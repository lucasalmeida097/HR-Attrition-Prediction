## 🛠️ Pré-processamento de Dados (Data Preprocessing)
1. Limpeza de Dados (Data Cleaning)  
2. Transformação de Dados (Data Transformation)  
3. Normalização/Padronização de Dados (Data Scaling)  


```python
# Variável alvo
target = "Attrition"

# Separar features e target
X = df.drop(columns=[target])
Y = df[target].map({'Yes': 1, 'No': 0})

# Separar colunas numéricas e categóricas
numeric_features = X.select_dtypes(include=['int64', 'float64']).columns.tolist()
categorical_features = X.select_dtypes(include=['object']).columns.tolist()

# Pré-processamento para variáveis numéricas
numeric_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='median')),
    ('scaler', StandardScaler())
])

# Pré-processamento para variáveis categóricas
categorical_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('onehot', OneHotEncoder(handle_unknown='igone', sparse_output=False))
])

#Combinar em um único transformador
preprocessor = ColumnTransformer(transformers=[
    ('num', numeric_transformer, numeric_features),
    ('cat', categorical_transformer, categorical_features)
])

#Separar treino e teste
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42, stratify=Y)

print(f'Dimensão do dataset: {df.shape}')
print(f"Treino: {X_train.shape}, Teste: {X_test.shape}")
```
